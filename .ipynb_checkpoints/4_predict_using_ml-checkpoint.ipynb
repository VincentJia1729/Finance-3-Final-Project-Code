# 4 Predict Using ML

We will now predict future volatility using each of the 4 ML models we fitted

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers import LSTM, Input, Dropout, Normalization, GlobalAveragePooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
import os.path
from keras.models import load_model
import matplotlib.pyplot as plt
```

```python
data_1_testing_x = np.load("./numpy_data/data_1_testing_x.npy")
data_1_training_x = np.load("./numpy_data/data_1_training_x.npy")
data_8_testing_x = np.load("./numpy_data/data_8_testing_x.npy")
data_8_training_x = np.load("./numpy_data/data_8_training_x.npy")
testing_y = np.load("./numpy_data/testing_y.npy")
training_y = np.load("./numpy_data/training_y.npy")
```

```python
dnn_model = load_model('ml_models/dnn_model.keras')
gew_dnn_model = load_model('ml_models/gew_dnn_model.keras')
lstm_model = load_model('ml_models/lstm_model.keras')
gew_lstm_model = load_model('ml_models/gew_lstm_model.keras')
```

```python
y_pred_dnn_model = dnn_model.predict(data_1_testing_x)
y_pred_gew_dnn_model = gew_dnn_model.predict(data_8_testing_x)
y_pred_lstm_model = lstm_model.predict(data_1_testing_x)
y_pred_gew_lstm_model = gew_lstm_model.predict(data_8_testing_x)
```

We will now graph the predicted volatility versus the realized volatility

```python
time = np.arange(len(testing_y))
```

```python
plt.plot(time, y_pred_dnn_model, label='Predicted DNN Volatility', color='blue', linestyle='-')
plt.plot(time, testing_y, label='Realized Volatility', color='red', linestyle='-')
plt.title("DNN Volatility Prediction v. Realized Volatility")
plt.xlabel("Time in Days")
plt.ylabel("Volatility")
```

```python
plt.plot(time, y_pred_gew_dnn_model, label='Predicted GEW DNN Volatility', color='blue', linestyle='-')
plt.plot(time, testing_y, label='Realized Volatility', color='red', linestyle='-')
plt.title("GEW DNN Volatility Prediction v. Realized Volatility")
plt.xlabel("Time in Days")
plt.ylabel("Volatility")
```

```python
plt.plot(time, y_pred_lstm_model, label='Predicted LSTM Volatility', color='blue', linestyle='-')
plt.plot(time, testing_y, label='Realized Volatility', color='red', linestyle='-')
plt.title("LSTM Volatility Prediction v. Realized Volatility")
plt.xlabel("Time in Days")
plt.ylabel("Volatility")
```

```python
plt.plot(time, y_pred_gew_lstm_model, label='Predicted GEW-LSTM Volatility', color='blue', linestyle='-')
plt.plot(time, testing_y, label='Realized Volatility', color='red', linestyle='-')
plt.title("GEW-LSTM Volatility Prediction v. Realized Volatility")
plt.xlabel("Time in Days")
plt.ylabel("Volatility")
```

Looking at the garphs, we can be fairly satisfied with our results! Now, we want to see which model has the lowest MSE over the time period.

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_dnn_model = mean_squared_error(y_pred_dnn_model, testing_y)
mse_gew_dnn_model = mean_squared_error(y_pred_gew_dnn_model, testing_y)
mse_lstm_model = mean_squared_error(y_pred_lstm_model,testing_y)
mse_gew_lstm_model = mean_squared_error(y_pred_gew_lstm_model, testing_y)
```

```python
print("Mean Squared Error for DNN Model (6 parameters) is: ", mse_dnn_model)
print("Mean Squared Error for GEW-DNN Model (14 parameters) is: ", mse_gew_dnn_model)
print("Mean Squared Error for LSTM Model (6 parameters) is: ", mse_lstm_model)
print("Mean Squared Error for GEW-LSTM Model (14 parameters) is: ", mse_gew_lstm_model)
```

**Dicussion of Results**  
Our expectation is that the GEW-LSTM Model should have the lowest MSE for the same amount of epochs among all the models. We can see this is the case. This also means that GEW-LSTM out performs the plain LSTM, which is expected. More parameters on the same model should perform better.   
We also see that the LSTM model is superior to the DNN model which is also expected. However, we see that the MSE of the GEW-DNN model is actually worse than the MSE of the DNN model. This is not expected, but the MSE values for both models are fairly close. We also see that the GEW-LSTM model is only slightly better than the LSTM model in terms of MSE.
